## las variables
1. CP --> costo del producto
2. CD --> Cantidad de dinero
3. Cambio


## formulas
$$Cambio = CD - CP$

# diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/IkerRivera/calcular-cambio-if-else/refs/heads/main/diagrma%20de%20flujo%20if%20else%20cambio%202.png" width="400" height="400" alt="Texto descriptivo">


```python
# para cuando es verdadero (si)
# calculando el cambio usando if else
# Datos de entrada CD, CP
# mostar cambio si es falso
# si es verdadero mostar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("programa realizado por iker montoya NL = 35")
```
